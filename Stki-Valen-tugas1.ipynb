#Masukkan berita kedalam string teks

```python
teks = """Bupati Purwakarta Dedi Mulyadi memimpin penggerebekan sebuah rumah milik seorang kuli bangunan Aman Permana (42), di Desa Cilegong, Kecamatan Jatiluhur, Kabupaten Purwakarta, Minggu (31/12/2017) malam. Penggerebekan tersebut dalam rangka kegiatan rutin Dedi menjelang tahun baru untuk memberikan bantuan kepada warganya yang kurang mampu. \"Saya kaget karena tiba-tiba rumah saya digerebek beberapa pria berbadan tegap dan berambut cepak meminta saya untuk menyerahkan anak saya,\" kata Aman kepada wartawan di lokasi kejadian. Rumah Aman terpilih sebagai target operasi penggerebekan bantuan setelah Dedi berkeliling selama tiga jam menyusuri gang-gang sempit di wilayah Jatiluhur. Beberapa pria berambut cepak menggedor pintu rumah Aman dan diminta menyerahkan diri. \"Setelah disuruh menyerahkan diri dan hendak dibawa ke kantor, saya langsung disuruh menghadap seseorang di depan pintu rumah. Ternyata Pak Bupati sudah menunggu dan memberi bantuan kepada saya dan keluarga. Aduh saya takut awalnya, tetapi bahagia akhirnya. Terima kasih, Pak,\" ujar Aman. Tahun Baru Terakhir sebagai Bupati, Dedi Mulyadi Layani Petugas Kebersihan Dedi menilai kegiatan rutinnya setiap malam tahun baru untuk memberikan sedikit kebahagiaan bagi warga kurang mampu. Ia menyebut, sudah ada peningkatan kesejahteraan dari tahun ke tahun. \"Sudah ada peningkatan sekarang, saya dari tadi nyari rumah reyot tiga jam keliling, baru ketemu. Sekarang susah nyari rumah reyot juga,\" kata Dedi. Dedi Mulyadi dan Ridwan Kamil Dianggap Kepala Daerah Terpegah 2017 Ia mengatakan, penggerebekan ini dilakukan secara spontan hasil keliling seusai mengikuti acara \"Tukar Nasib\". \"Alhamdulilah program pemerintah tersebut dilaksanakan secara tepat sasaran,\" ujarnya."""
teks
```

#Tokenisasi menggunakan nltk(neutural language)

```python
tokens = nltk.word_tokenize(teks)
tokens
```

#Menghapus karakter khusus

```python
tokens = [token for token in tokens if token not in string.punctuation]
tokens
```

#Mengubah teks menjadi huruf kecil (lowercase)

```python
tokens = [token.lower() for token in tokens]
tokens
```

#Menghapus stopwords

```python
stop_words = set(stopwords.words('indonesian'))
tokens = [token for token in tokens if token not in stop_words]
tokens
```

```python
teks_preprocessed = ' '.join(tokens)

print(teks_preprocessed)
```
